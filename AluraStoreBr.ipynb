### Importação dos dados



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = sum(loja1['Preço'])
faturamento_loja2 = sum(loja2['Preço'])
faturamento_loja3 = sum(loja3['Preço'])
faturamento_loja4 = sum(loja4['Preço'])

print(f"Faturamento total da loja 1: R${faturamento_loja1:,.2f}")
print(f"Faturamento total da loja 2: R${faturamento_loja2:,.2f}")
print(f"Faturamento total da loja 3: R${faturamento_loja3:,.2f}")
print(f"Faturamento total da loja 4: R${faturamento_loja4:,.2f}")
```

# 2. Vendas por Categoria


```python
vendas_categoria_loja1 = loja1.groupby('Categoria do Produto').size()
vendas_categoria_loja2 = loja2.groupby('Categoria do Produto').size()
vendas_categoria_loja3 = loja3.groupby('Categoria do Produto').size()
vendas_categoria_loja4 = loja4.groupby('Categoria do Produto').size()
```

# 3. Média de Avaliação das Lojas

```python
media_loja1 = np.mean(loja1['Avaliação da compra'])
media_loja2 = np.mean(loja2['Avaliação da compra'])
media_loja3 = np.mean(loja3['Avaliação da compra'])
media_loja4 = np.mean(loja4['Avaliação da compra'])

media_loja1
```

```python
avaliacoes_loja1 = loja1.groupby('Avaliação da compra').size()
avaliacoes_loja2 = loja2.groupby('Avaliação da compra').size()
avaliacoes_loja3 = loja3.groupby('Avaliação da compra').size()
avaliacoes_loja4 = loja4.groupby('Avaliação da compra').size()

avaliacoes_loja1
```

# 4. Produtos Mais e Menos Vendidos

```python
vendas_produto_loja1 = loja1.groupby('Produto').size()
vendas_produto_loja2 = loja2.groupby('Produto').size()
vendas_produto_loja3 = loja3.groupby('Produto').size()
vendas_produto_loja4 = loja4.groupby('Produto').size()
```

# 5. Frete Médio por Loja


```python
frete_medio_loja1 = np.mean(loja1['Frete'])
frete_medio_loja2 = np.mean(loja2['Frete'])
frete_medio_loja3 = np.mean(loja3['Frete'])
frete_medio_loja4 = np.mean(loja4['Frete'])

fretes_medios = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]

fretes_loja1 = loja1['Frete']
fretes_loja2 = loja2['Frete']
fretes_loja3 = loja3['Frete']
fretes_loja4 = loja4['Frete']

fretes = [fretes_loja1, fretes_loja2, fretes_loja3, fretes_loja4]
```

# Relatório Final

A análise que será feita aqui pretende indicar qual loja deve ser vendida com base nos dados extraídos, comparando o desempenho de cada loja. No final, será indicado qual loja deve ser vendida a partir da análise realizada

## Gráficos

### Gráfico de análise de vendas por categoria

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

vendas_categoria_lojas = [vendas_categoria_loja1, vendas_categoria_loja2, vendas_categoria_loja3, vendas_categoria_loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

cores = ['#015BA0', '#0077B6', '#0096C7', '#00B4D8', '#48CAE4', '#90E0EF', '#ADE8F4', '#CAF0F8']

for ax, vendas_categoria, nome in zip(axes.flatten(), vendas_categoria_lojas, nomes_lojas):
    ax.pie(
        vendas_categoria,
        labels=vendas_categoria.index,
        autopct='%1.1f%%',
        startangle=90,
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'},
        colors=cores

    )
    ax.set_title(f'Vendas por Categoria - {nome}')

plt.tight_layout()
plt.show()
```

### Gráfico da média de avaliação das lojas

```python
fig2, axes2 = plt.subplots(2, 2, figsize=(10, 8))

avaliacoes_lojas = [avaliacoes_loja1, avaliacoes_loja2, avaliacoes_loja3, avaliacoes_loja4]
medias_lojas = [media_loja1, media_loja2, media_loja3, media_loja4]

for ax, avaliacao, media, nome in zip(axes2.flatten(), avaliacoes_lojas, medias_lojas, nomes_lojas):
    ax.stem(avaliacao.index, avaliacao, basefmt='--')
    ax.set_title(f"{nome} - Média {media:.2f}")
    ax.set_xlabel('Avaliação')
    ax.set_ylabel('Quantidade de avaliações')
    ax.set_xticks(range(1, 6, 1))

plt.tight_layout()
plt.show()
```

### Gráfico de vendas dos produtos

```python
fig3, axes3 = plt.subplots(4, 1, figsize=(10, 16))

vendas_produtos = [vendas_produto_loja1, vendas_produto_loja2, vendas_produto_loja3, vendas_produto_loja4]

for ax, vendas_produto, nome in zip(axes3, vendas_produtos, nomes_lojas):
    menos_vendido = vendas_produto.idxmin()
    qtd_menor = vendas_produto.min()
    mais_vendido = vendas_produto.idxmax()
    qtd_maior = vendas_produto.max()
    ax.bar(vendas_produto.index, vendas_produto)
    ax.set_title(
    f'Vendas por Produto\n'
    f'Menos vendido: {menos_vendido} ({qtd_menor}) | Mais vendido: {mais_vendido} ({qtd_maior})',
    fontsize=11,
    fontweight='bold'
    )
    ax.set_ylabel('Quantidade de vendas')
    ax.set_yticks(range(0, qtd_maior + 10, 10))
    ax.tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

### Gráfico dos fretes médios das lojas

```python
fig4, axes4 = plt.subplots(2, 2, figsize=(10, 8))

for ax, frete_medio, frete, nome in zip(axes4.flatten(), fretes_medios, fretes, nomes_lojas):
    ax.scatter(frete.index, frete.values, color='teal', alpha=0.7)
    ax.set_title(f'{nome} - Frete Médio: {frete_medio:.2f}')
    ax.set_xlabel('Índice do pedido')
    ax.set_ylabel('Valor do Frete')

plt.tight_layout()
plt.show()
```

## Conclusão dos gráficos

Os gráficos expostos anteriormente expõe que as lojas são bem semelhantes quando falamos sobre vendas por categoria, média de avaliação e frete médio. Existe uma discrepância entre as lojas quando falamos sobre os produtos mais vendidos em cada uma, entretanto isso não deve influenciar na análise de desempenho de cada loja, visto que apesar de terem diferentes produtos como mais/menos comprados entre si, as categorias mais buscadas em cada loja é muito semelhante. Levando tudo isso em consideração, devemos olhar para o aspecto que aparenta ser o mais relevante: o faturamento de cada loja.

### Gráfico de faturamento

```python
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
faturamentos_milhao = [f / 1_000_000 for f in faturamentos]

cores_faturamento = ['#001219', '#005F73', '#0A9396', '#94D2BD']

plt.bar(nomes_lojas, faturamentos_milhao, color=cores_faturamento)
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (em Milhão de R$)')
plt.show()
```

Podemos ver então, que mesmo as lojas possuindo dados bem semelhantes, ainda assim elas possuem faturamentos diferentes. Sendo assim, a loja que deve ser vendida é a loja 4, já que mesmo sendo semelhante às outras lojas ela ainda possui um faturamento menos que as outras.
